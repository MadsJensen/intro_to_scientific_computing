### Super User / root priviledge

To make changes to the VM, such as installing software packages, you will need to have _administrative priviledges_. In Linux-speak, the 'Administrator' is known as `root` (not to be confused with the root of the file system, `/`). In many ([Ubuntu](https://www.ubuntu.com/)-based) distributions like [Mint](https://www.linuxmint.com/), the normal user account created during setup (in our VM, simply: `user`) gets so-called ___super user priviledge___. This simply means that by prepending the word `sudo` to any command will _temporarily elevate_ `user`s priviledge to that or `root`, _e.g._

```sudo apt-get install cowsay```

# Conda and Anaconda

## What is `conda`?

From the [project webpage](https://conda.io/docs/):

>Conda is an open source ___package___ management system ___and environment management___ system that runs on Windows, macOS and Linux. Conda quickly installs, runs and updates packages and their ___dependencies___. Conda easily creates, saves, loads and switches between environments on your local computer. It was created for Python programs, but it can package and distribute software for any language.

_Keywords: package; environment; management system; dependencies_

## Why do we want it?

We are learning about computing, for which we will need to write [programs](computer-programs-and-data-types.ipynb). The specific language we will use for demonstrations and exercises is called [Python](https://en.wikipedia.org/wiki/Python_(programming_language). We use `conda` to ease the pain of installing and operating a Python-based programming environment on all 3 OS's. Much of the said 'pain' arises due to _interdependencies_: some software modules/packages _depend_ on others being present. This can quickly become very complicated without a tool like `conda`...

We've created an _environment_-definition, containing all the tools we'll need during this course (we'll discuss what this entails later). The Linux VM already has the environment installed and _activated_ by default.

## Quitting JupyterLab

Closing the browser window will not close the _notebook server_. Correspondingly, if you accidentally close the browser window/tab `JupyterLab` is in, you can return to it by going to the appropriate URL. The notebook server prints out the URL on startup (see the Prompt/Terminal output right after starting). By default, the first server you start will be found at

```
http://localhost:8888
```

To _really_ quit the server, either close the Prompt/Terminal window, or hit Ctrl-C _twice_ in the terminal.
